```python
import pandas as pd
print(pd.__version__)
```

```python
df = pd.DataFrame({'Birth_Str':['2019-01-01 09:10:00',
                            '2019-01-08 09:20:30',
                            '2019-02-01 10:20:00',
                            '2019-02-02 11:40:50',
                            '2019-02-28 15:10:20',
                            '2019-04-10 19:20:50',
                            '2019-06-30 21:20:50',
                            '2019-07-20 23:30:59']})
df
```

```python
df['Birth_DT'] = pd.to_datetime(df['Birth_Str'], format='%Y-%m-%d %H:%M:%S', errors='raise')
df
```

| 포맷 | 의미 | 예 |
|:---|:------|:------|
| %Y | Year | 2019, 2020 |
| %m | Month as a zero-padded | 01~12 |
| %d | Day of the month as a zero-padded | 01~31 |
| %H | Hour (24-hour clock) as a zero-padded | 01~23 |
| %M | Minute as a zero-padded | 00~59 |
| %S | Second as a zero-padded | 00~59 |



* **2019-09-01 19:30:00** => <font color=red>**%Y-%m-%d %H:%M:%S**</font>

```python
df['Birth_date']       = df['Birth_DT'].dt.date         # YYYY-MM-DD(문자)
df['Birth_year']       = df['Birth_DT'].dt.year         # 연(4자리숫자)
df['Birth_month']      = df['Birth_DT'].dt.month        # 월(숫자)
df['Birth_month_name'] = df['Birth_DT'].dt.month_name() # 월(문자)

df['Birth_day']        = df['Birth_DT'].dt.day          # 일(숫자)
df['Birth_time']       = df['Birth_DT'].dt.time         # HH:MM:SS(문자)
df['Birth_hour']       = df['Birth_DT'].dt.hour         # 시(숫자)
df['Birth_minute']     = df['Birth_DT'].dt.minute       # 분(숫자)
df['Birth_second']     = df['Birth_DT'].dt.second       # 초(숫자)

df
```

```python
df['Birth_quarter']       = df['Birth_DT'].dt.quarter       # 분기(숫자)
df['Birth_weekday_name']  = df['Birth_DT'].dt.day_name()  # 요일이름(문자) (=day_name())
df['Birth_weekday']       = df['Birth_DT'].dt.weekday       # 요일숫자(0-월, 1-화) (=dayofweek)
df['Birth_weekofyear']    = df['Birth_DT'].dt.weekofyear    # 연 기준 몇주째(숫자) (=week)
df['Birth_dayofyear']     = df['Birth_DT'].dt.dayofyear     # 연 기준 몇일째(숫자)
df['Birth_days_in_month'] = df['Birth_DT'].dt.days_in_month # 월 일수(숫자) (=daysinmonth)


df
```

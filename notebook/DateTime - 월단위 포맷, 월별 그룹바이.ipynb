```python
names = ['ID', 'DATE', 'TIMESTAMP', 'VAL1', 'VAL2', 'VAL3', 'VAL4']
df = pd.read_csv('./data/2020-04-20-Monthly-Sum-Data.csv', encoding='utf-8', names=names)
print(df.shape)
df.head()
```

### Timestamp 포맷의 문자열을 Datetime 데이터 타입으로 변경

```python
df['TIMESTAMP'] = pd.to_datetime(df['TIMESTAMP'])
```

```python
df.head()
```

```python
df['date_minus_time'] = df["TIMESTAMP"].apply(lambda df: datetime.datetime(year=df.year, month=df.month, day=df.day))
df
```

```python
df.set_index(df["date_minus_time"],inplace=True)
df
```

```python
df.resample('M').sum()
```

### 'YYYYMMDD' 포맷의 문자열을 DateTime 데이터 타입으로 변경

```python
df['DATE'] = df["DATE"].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
df
```

```python
df.info()
```

```python
df.set_index(df["DATE"],inplace=True)
df
```

# DateTime을 YYYY-MM 포맷으로 변경하기

```python
df['yr_mnth'] = df['TIMESTAMP'].apply(lambda x: x.strftime('%Y-%m'))
df
```

```python
df.groupby('yr_mnth')['VAL1'].mean()
```
